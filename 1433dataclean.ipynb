```python
from numpy.lib.function_base import diff
import pandas as pd

filename = "/Users/jasonlee/Downloads/1433longdata.csv"
df = pd.read_csv(filename)
dic =  {'ID': [], 'School': [], 'Year': [], 'Policy': [], 'TotalApps': [], 'TotalAdds': [], 'SATR75': [], 'SATR25': [], 'SATM75': [], 'SATM25': [],  'SATrate': [], 'PellRate': [], 'TotalStudents': [], 'TotalWomen': [], 'TotalAmInd': [], 'TotalAsian': [], 'TotalBlack': [], 'TotalHispanic': [], 'TotalPacIsland': [], 'TotalWhite': [], 'TotalTwo': [], 'AdmitRate': [], 'AdmitYield': [], 'Level': [], 'Sector': [], 'State': [], 'Urban': [], 'PriceInState': [], 'PriceOutState': []}

year = 2021
i = 0
numRows = df.shape[0]
numRows1 = df.shape[0]
numRows2 = df.shape[0]
numRows3 = df.shape[0]
numRows4 = df.shape[0]
numRows5 = df.shape[0]
numRows6 = df.shape[0]
numRows7 = df.shape[0]
numRows8 = df.shape[0]
numRows9 = df.shape[0]

while year != 2015:
    for column in df:
        for i in df[column].items():
            if column == "Institution Name":
                dic['School'].append(i[1])
            if column == "UnitID":
                dic['ID'].append(i[1])
            if column[:21] == "Admission test scores" and column[len(column)-5: len(column)-1] == str(year):
                dic['Year'].append(year)
                dic['Policy'].append(i[1])
            if column[:16] == "Applicants total" and column[len(column)-5: len(column)-1] == str(year):
                dic['TotalApps'].append(i[1])
            if column[:16] == "Admissions total" and column[len(column)-5: len(column)-1] == str(year):
                dic['TotalAdds'].append(i[1])
            if (column[:43] == "SAT Evidence-Based Reading and Writing 75th" or column[:42] == "SAT Critical Reading 75th percentile score") and column[len(column)-5: len(column)-1] == str(year):
                dic['SATR75'].append(i[1])
            if (column[:43] == "SAT Evidence-Based Reading and Writing 25th" or column[:42] == "SAT Critical Reading 25th percentile score") and column[len(column)-5: len(column)-1] == str(year):
                dic['SATR25'].append(i[1])
            if column[:30] == "SAT Math 75th percentile score" and column[len(column)-5: len(column)-1] == str(year):
                dic['SATM75'].append(i[1])
            if column[:30] == "SAT Math 25th percentile score" and column[len(column)-5: len(column)-1] == str(year):
                dic['SATM25'].append(i[1])
            if column[:48] == "Percent of first-time degree/certificate-seeking" and column[len(column)-5: len(column)-1] == str(year): 
                dic['SATrate'].append(i[1])
            if column[:66] == "Percent of full-time first-time undergraduates awarded Pell grants":
                if year == 2021 and numRows != 0:
                    dic['PellRate'].append(-9999)
                    numRows -= 1
                elif column[len(column)-5: len(column)-3] == str(year)[2:]: 
                    dic['PellRate'].append(i[1])  
            if column[:11] == "Grand total":
                if year == 2021 and numRows1 != 0:
                    dic['TotalStudents'].append(-9999)
                    numRows1 -= 1
                elif column[15:19] == str(year):
                    dic['TotalStudents'].append(i[1]) 
            if column[:11] == "Total women":
                if year == 2021 and numRows2 != 0:
                    dic['TotalWomen'].append(-9999)
                    numRows2 -= 1
                elif column[15:19] == str(year):
                    dic['TotalWomen'].append(i[1])
            if column[:38] == "American Indian or Alaska Native total":
                if year == 2021 and numRows3 != 0:
                    dic['TotalAmInd'].append(-9999)
                    numRows3 -= 1
                elif column[42:46] == str(year):
                    dic['TotalAmInd'].append(i[1]) 
            if column[:11] == "Asian total":
                if year == 2021 and numRows4 != 0:
                    dic['TotalAsian'].append(-9999)
                    numRows4 -= 1
                elif column[15:19] == str(year):
                    dic['TotalAsian'].append(i[1]) 
            if column[:31] == "Black or African American total":
                if year == 2021 and numRows5 != 0:
                    dic['TotalBlack'].append(-9999)
                    numRows5 -= 1
                elif column[35:39] == str(year):
                    dic['TotalBlack'].append(i[1]) 
            if column[:14] == "Hispanic total":
                if year == 2021 and numRows6 != 0:
                    dic['TotalHispanic'].append(-9999)
                    numRows6 -= 1
                elif column[18:22] == str(year):
                    dic['TotalHispanic'].append(i[1]) 
            if column[:47] == "Native Hawaiian or Other Pacific Islander total":
                if year == 2021 and numRows7 != 0:
                    dic['TotalPacIsland'].append(-9999)
                    numRows7 -= 1
                elif column[51:55] == str(year):
                    dic['TotalPacIsland'].append(i[1]) 
            if column[:11] == "White total":
                if year == 2021 and numRows8 != 0:
                    dic['TotalWhite'].append(-9999)
                    numRows8 -= 1
                elif column[15:19] == str(year):
                    dic['TotalWhite'].append(i[1]) 
            if column[:23] == "Two or more races total":
                if year == 2021 and numRows9 != 0:
                    dic['TotalTwo'].append(-9999)
                    numRows9 -= 1
                elif column[27:31] == str(year):
                    dic['TotalTwo'].append(i[1]) 
            if column[:16] == "Percent admitted" and column[len(column)-5: len(column)-1] == str(year): 
                dic['AdmitRate'].append(i[1])
            if column[:16] == "Admissions yield" and column[len(column)-5: len(column)-1] == str(year): 
                dic['AdmitYield'].append(i[1])
            if column[:20] == "Level of institution" and column[len(column)-5: len(column)-1] == str(year): 
                dic['Level'].append(i[1])
            if column[:21] == "Sector of institution" and column[len(column)-5: len(column)-1] == str(year): 
                dic['Sector'].append(i[1])
            if column[:18] == "State abbreviation" and column[len(column)-5: len(column)-1] == str(year): 
                dic['State'].append(i[1])
            if column[:45] == "Degree of urbanization (Urban-centric locale)" and column[len(column)-5: len(column)-1] == str(year): 
                dic['Urban'].append(i[1])
            if column[:50] == "Total price for in-state students living on campus" and column[len(column)-5: len(column)-1] == str(year):
                dic['PriceInState'].append(i[1])
            if column[:54] == "Total price for out-of-state students living on campus" and column[len(column)-5: len(column)-1] == str(year):
                dic['PriceOutState'].append(i[1])
    
    year -= 1
    numRows = df.shape[0]
    numRows1 = df.shape[0]
    numRows2 = df.shape[0]
    numRows3 = df.shape[0]
    numRows4 = df.shape[0]
    numRows5 = df.shape[0]
    numRows6 = df.shape[0]
    numRows7 = df.shape[0]
    numRows8 = df.shape[0]
    numRows9 = df.shape[0]

result = pd.DataFrame(dic)
result.to_csv('/Users/jasonlee/Downloads/1433longdataR.csv')


```
